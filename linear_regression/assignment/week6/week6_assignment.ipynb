```python
sale=pd.read_csv('kc_house_data_small.csv',dtype=dtype_dict)
test=pd.read_csv('kc_house_data_small_test.csv',dtype=dtype_dict)
train=pd.read_csv('kc_house_data_small_train.csv',dtype=dtype_dict)
validation=pd.read_csv('kc_house_data_validation.csv',dtype=dtype_dict)
```

```python
def get_numpy_data(data,features,output):
    data['constant'] = 1
    features = ['constant'] + features
    featuredata=data[features]
    features_matrix=featuredata.values
    outdata=data[output]
    output_array =outdata.values
    return(features_matrix, output_array)
```

```python
def normalize_features(features):              #input is this is for all the features!
    norms=np.linalg.norm(features,axis=0)
    normalized_features=features/norms
    return (normalized_features, norms)
```

```python
def k_nearest_neighbors(k,feature_train,feature_query):
        diff=feature_train-feature_query
        distance=np.sqrt(np.sum(diff**2,axis=1))
        index=np.argsort(distance)
        tu=[]
        for i in range(0,k):
            tu.append(index[i])
        return tu
# return list of different k nearest neighbors.
```

```python
k_nearest_neighbors(4,features_train,query)
```

```python
all=[]
for i in k_nearest_neighbors(4,features_train,query):
    all.append(train.iloc[i].price)
np.mean(all)
```

```python
def predict(k,feature_train,output_train,query):
    klist=k_nearest_neighbors(k,features_train,query)
    all=[]
    for i in klist:
        all.append(output_train[i])
    return np.mean(all)
```

```python
predict(10,features_train,output_train,query)
```
